# 00 — Import All Public Modules

This notebook is a **smoke test**: it imports every public module in AbaQuant
and confirms the package loads cleanly. It is a good first notebook to run
after installing AbaQuant, since it will fail fast if something is missing
from your environment.

**What you'll see:**
- How to make `abaquant` importable from a notebook (works both for an
  installed package and for a local source checkout).
- The full list of public modules covered by AbaQuant's stable v1 namespace.


## Setup

Make sure `abaquant` can be imported, whether it's installed or you're running from a local source checkout.


```python
import os, sys
from pathlib import Path

os.environ.setdefault("MPLBACKEND", "Agg")

def _ensure_abaquant_importable():
    try:
        import abaquant  # noqa: F401
        return
    except ModuleNotFoundError:
        pass
    here = Path.cwd()
    for candidate in [here, *here.parents]:
        src = candidate / "src"
        if (src / "abaquant" / "__init__.py").exists():
            sys.path.insert(0, str(src))
            return
    raise ModuleNotFoundError(
        "Could not import abaquant. Run this notebook from within the AbaQuant "
        "repository (or pip install abaquant)."
    )

_ensure_abaquant_importable()
import abaquant
print(f"AbaQuant version: {abaquant.__version__}")
```

## Run the import smoke test


```python
def import_public_modules():
    """Import every public module and return its import status."""
    return {
        module_name: type(importlib.import_module(module_name)).__name__
        for module_name in PUBLIC_MODULES
    }

statuses = import_public_modules()
print(f"Imported {len(statuses)} modules successfully.\n")
for name, kind in list(statuses.items())[:8]:
    print(f"{name:45s} -> {kind}")
```
